So now that we have the matrix, what can we do with it?

One initial metric we an obtain is the indegree and outdegree of each node.

We'll define outdegree as the number of entries that a given entry (when treated as a substring) maps on to.
We'll define indegree as the number of entries (when treated as a substring) which map on to given entry.

Note, that this is non symetric relationship.  By this, we mean that if entry A is a substring of B it does not follow that B is a substring of A (indeed, A would only be a substring of B in the case of identity, where A=B).  As such this the graph/network we derive from this process is *directed*

NOTE: given the size of the matrix being used in this computation (even if it is a sparse matrix), we'll try and use an existing file if it exists

```python
import numpy as np

inOutDegreePath=os.path.join(gitDir,'ossPy/PackageOuts/inOutDegree.csv')

#load it if it exists, otherwise make it
if os.path.isfile(matrixPath):
    uniqueNamesAndCounts=pd.read_csv(inOutDegreePath)
    uniqueNamesAndCounts.drop(columns=['Unnamed: 0'])
else:

    #establish the table columns
    uniqueNamesAndCounts['outdegree']=0
    uniqueNamesAndCounts['indegree']=0

    #convert to manipulable format
    accessibleMatrix=accessibleMatrix.tocsr()

    #iterate across nodes
    for index, row in uniqueNamesAndCounts.iterrows():
    
        #compute indegree and outdegree
        uniqueNamesAndCounts['outdegree'].loc[index]=accessibleMatrix[index,:].count_nonzero
        uniqueNamesAndCounts['indegree'].loc[index]=accessibleMatrix[:,index].count_nonzero
```

What we see here are entries which are mapping on to extremely large numbers of other entries.  In essence, because of how vague these terms are, they are mapping on to these large numbers of other terms, and are thus uninformative.  We should note that outdegree *alone* isn't sufficient for making this determination.  For example, when we look at 'student' or 'microsoft' we see that there are a number of entries that these terms map on to (due to the variability in how people are expressing these affiliations), but there's still high count of individuals associated with these terms.  This suggests that a ratio of the count and the outdegree may be a more informative measure of being "overly vague"
